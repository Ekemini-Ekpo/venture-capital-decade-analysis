## Data Cleaning and Preparation

### 1. Check for Null Values

```python
print('Null values in df_rounds_old:')
display(df_rounds_old.isnull().sum()[df_rounds_old.isnull().sum() > 0])

print('\nNull values in df_rounds_new:')
display(df_rounds_new.isnull().sum()[df_rounds_new.isnull().sum() > 0])

print('\nNull values in df_startups_new:')
display(df_startups_new.isnull().sum()[df_startups_new.isnull().sum() > 0])
```

### 2. Drop rows with null `raised_amount_usd` in `df_rounds_old`

```python
initial_rows_rounds_old = df_rounds_old.shape[0]
df_rounds_old.dropna(subset=['raised_amount_usd'], inplace=True)
rows_dropped_rounds_old = initial_rows_rounds_old - df_rounds_old.shape[0]
print(f"Dropped {rows_dropped_rounds_old} rows from df_rounds_old due to null 'raised_amount_usd'.")
```

### 3. Drop rows with null `amount_million_usd` in `df_rounds_new`

```python
initial_rows_rounds_new = df_rounds_new.shape[0]
df_rounds_new.dropna(subset=['amount_million_usd'], inplace=True)
rows_dropped_rounds_new = initial_rows_rounds_new - df_rounds_new.shape[0]
print(f"Dropped {rows_dropped_rounds_new} rows from df_rounds_new due to null 'amount_million_usd'.")
```

### 4. Convert `raised_amount_usd` to `amount_million_usd` in `df_rounds_old`

```python
df_rounds_old['amount_million_usd'] = df_rounds_old['raised_amount_usd'] / 1_000_000
print("Created 'amount_million_usd' column in df_rounds_old.")
display(df_rounds_old[['raised_amount_usd', 'amount_million_usd']].head(3))
```

### 5. Extract year from `funded_at` in `df_rounds_old`

```python
df_rounds_old['funded_at'] = pd.to_datetime(df_rounds_old['funded_at'], errors='coerce')
df_rounds_old['year'] = df_rounds_old['funded_at'].dt.year
print("Extracted 'year' from 'funded_at' in df_rounds_old.")
display(df_rounds_old[['funded_at', 'year']].head(3))
```

### 6. Print shapes after cleaning

```python
print('Shape of df_rounds_old after cleaning:', df_rounds_old.shape)
print('Shape of df_rounds_new after cleaning:', df_rounds_new.shape)
print('Shape of df_startups_new (unchanged):', df_startups_new.shape)
```

## Data Visualizations

### 2. Total Funding Amount per Year from `df_rounds_old` (2004 to 2013)

```python
df_rounds_old_filtered = df_rounds_old[(df_rounds_old['year'] >= 2004) & (df_rounds_old['year'] <= 2013)]
total_funding_per_year = df_rounds_old_filtered.groupby('year')['amount_million_usd'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='amount_million_usd', data=total_funding_per_year, marker='o')
plt.title('Total Funding Amount per Year (2004-2013) - df_rounds_old')
plt.xlabel('Year')
plt.ylabel('Total Funding Amount (Million USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 6. Then vs Now: How Funding Stage Distribution Has Shifted (2004-2013 vs 2020-2026)

```python
# Get top 5 funding round types from df_rounds_old
top_5_rounds_old = df_rounds_old['funding_round_type'].value_counts().nlargest(5).reset_index()
top_5_rounds_old.columns = ['funding_stage', 'count']
top_5_rounds_old['Source'] = '2004-2013'

# Get top 5 funding stages from df_rounds_new
top_5_rounds_new = df_rounds_new['funding_stage'].value_counts().nlargest(5).reset_index()
top_5_rounds_new.columns = ['funding_stage', 'count']
top_5_rounds_new['Source'] = '2020-2026'

# Combine the dataframes
combined_funding_stages = pd.concat([top_5_rounds_old, top_5_rounds_new])

plt.figure(figsize=(14, 7))
sns.barplot(x='funding_stage', y='count', hue='Source', data=combined_funding_stages,
            palette=['#4c72b0', '#dd8452']) # Using seaborn's default blue and orange for contrast

plt.title('Then vs Now: How Funding Stage Distribution Has Shifted (2004-2013 vs 2020-2026)')
plt.xlabel('Funding Stage Type')
plt.ylabel('Number of Deals')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Time Period')
plt.tight_layout()
plt.show()
```

## Historical Funding Stage Distribution (2004-2013)

```python
# Filter df_rounds_old for specific funding round types
specific_funding_types = ['seed', 'series-a', 'series-b', 'series-c']
df_filtered = df_rounds_old[df_rounds_old['funding_round_type'].isin(specific_funding_types)]

# Count deals by funding_round_type and sort
deal_counts = df_filtered['funding_round_type'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12, 7))
ax = sns.barplot(x=deal_counts.values, y=deal_counts.index, palette='Blues_d')

plt.title('Historical Funding Stage Distribution (2004-2013)')
plt.xlabel('Number of Deals')
plt.ylabel('Funding Round Type')

# Add data labels
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 50, # Add a small offset to the right of the bar
             p.get_y() + p.get_height() / 2,
             '{:1.0f}'.format(width),
             ha='left', va='center')

plt.tight_layout()
plt.show()
```

## Modern Funding Stage Distribution (2020-2026)

```python
# Count deals by funding_stage and sort
deal_counts_new = df_rounds_new['funding_stage'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12, 7))
ax = sns.barplot(x=deal_counts_new.values, y=deal_counts_new.index, palette='Oranges_d')

plt.title('Modern Funding Stage Distribution (2020-2026)')
plt.xlabel('Number of Deals')
plt.ylabel('Funding Stage')

# Add data labels
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 20, # Add a small offset to the right of the bar
             p.get_y() + p.get_height() / 2,
             '{:1.0f}'.format(width),
             ha='left', va='center')

plt.tight_layout()
plt.show()
```

Note: The historical dataset contains investor-type classifications (venture, angel) alongside stage classifications. This analysis isolates true stage-based rounds for a clean comparison.

## Historical Stage-Based Funding Distribution (2004-2013) - Filtered Stages

```python
# Filter df_rounds_old for specific stage-based funding round types
specific_stage_types = ['series-a', 'series-b', 'series-c+']
df_stage_filtered = df_rounds_old[df_rounds_old['funding_round_type'].isin(specific_stage_types)]

# Count deals by funding_round_type and sort
deal_counts_stage = df_stage_filtered['funding_round_type'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
ax = sns.barplot(x=deal_counts_stage.values, y=deal_counts_stage.index, palette='Blues_d')

plt.title('Historical Stage-Based Funding Distribution (2004-2013)', fontsize=14)
plt.xlabel('Number of Deals', fontsize=12)
plt.ylabel('Funding Round Type', fontsize=12)

# Add data labels
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 50, # Add a small offset to the right of the bar
             p.get_y() + p.get_height() / 2,
             '{:1.0f}'.format(width),
             ha='left', va='center')

plt.tight_layout()
plt.show()
```

The center of gravity in venture capital moved earlier by a full stage over the decade. This reflects lower startup costs, faster iteration cycles, and investor preference for portfolio volume over concentrated late-stage bets


## **Key Insights**

1. Series A was the dominant entry point for institutional capital in 2004-2013, representing nearly double the deal volume of Series B and C+ combined. Investors bet on companies with demonstrated traction.
2. By 2020-2026 the market shifted dramatically earlier, Seed and Pre-Seed now account for the majority of deals, reflecting lower startup costs and investor appetite for earlier, higher-volume bets.
3. AI and Fintech have emerged as the defining sectors of the modern era, displacing the broader sector distribution seen historically.
4. The United States maintains disproportionate dominance in global startup activity even in the modern dataset, suggesting geographic concentration persists despite globalization.

Note: The historical dataset classifies some deals by investor type (venture, angel) rather than company stage, this analysis isolates only true stage-based rounds for a valid comparison
